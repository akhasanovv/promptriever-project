```python
dim = 768
batch_size = 64
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("intfloat/multilingual-e5-base")
```

```python
!pip install hnswlib
```

```python
from tqdm import tqdm
import numpy as np
```

```python
unique_context_cnt = 9080

embeddings = np.zeros((unique_context_cnt, dim))
contexts = []
idxs = np.arange(len(train_data))

previous_context = ""
cur_idx = 0

for (index, row) in tqdm(enumerate(train_data)):
    if row['context'] == previous_context:
        continue

    previous_context = row['context']
    passage_context = f"passage: {row['context']}"

    if index < 10:
        print(passage_context)

    contexts.append(previous_context)
    embeddings[cur_idx] = model.encode(passage_context)
    cur_idx += 1
```

## Собираем поисковый индекс (на основе алгоритма HNSW)

```python
import hnswlib
import pickle
```

```python
p = hnswlib.Index(space='cosine', dim=dim)
```

```python
max_elements = 10000
p.init_index(max_elements=max_elements, ef_construction=300, M=50)
```

```python
idxs = np.arange(embeddings.shape[0])
p.add_items(embeddings, idxs)
```

```python
def my_vector_search(query: str, top_k: int = 1):
    query_embedding = model.encode(f"query: {query}")

    labels, distances = p.knn_query(query_embedding, k=top_k)

    res_contexts = []
    for j in labels[0]:
        res_contexts.append(contexts[j])

    return (res_contexts, distances)
```

```python
query = "Какой коворкинг предоставляет рабочие места писателям?"
print(my_vector_search(query))
```

Как можно заметить, большое значение здесь имеют параметры $EF$ и $M$. Происходит tradeoff скорости / качества поиска.
